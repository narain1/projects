```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/t_bg_wf.csv').set_index('id')
np.random.seed(10)
alpha = np.random.dirichlet(np.ones(11), size=1, ).squeeze()
groups = dict()
k = 1000
ks = df.sum(axis=0).values/alpha
ks_idx = ks.argsort()
df['column_sum'] = df.sum(axis=1)
df.sort_values(by='column_sum', inplace=True)
```

```python
for idx, k_id in enumerate(ks_idx):
    temp = df
    n_select = int(alpha[k_id] * k)
    select_idxs = np.all([(temp[f'wf_{ks_idx[idx]+1}'] == 1).values, (temp['column_sum'] >=1).values], axis=0)
    vals = temp.iloc[select_idxs, :-1].head(n_select)
    groups[f'omega_{k_id}'] = vals
    print(f'omega_{k_id}', vals.shape)
    df = temp.drop(vals.index)
    print(df.shape, vals.shape)
```
