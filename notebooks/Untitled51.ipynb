```python
import spacy, random


nlp = spacy.load('en')
```

```python
def find_similar4(content, library, count, sentence_correction=False):
    acc = []
    nouns = [noun_chunk.text for noun_chunk in nlp(content) if noun_chunk.pos_ == 'NOUN']
    for _ in range(count):
        item = random.choice(library)
        o = ' '.join([i.text if i.pos_ != 'NOUN' else random.choice(nouns) for i in nlp(item)])
        if sentence_correction == 'true':
            o = list(gf.correct(o, max_candidates=20))[0] if random.random() > 0.9 else o
        acc.append(o)
    return {'generated_values': acc[:count]}
```
